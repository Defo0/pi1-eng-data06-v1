voy a crear 3 listas, una con los dataframes, otra con las plataformas y otra con las iniciales, las cuales van a ser utilizadas posteriormente en las distintas fucniones de transformacion. las creo ahora y en orden para no olvidarme despues y ya queden hechas. 

```python
# Crear una lista de DataFrames, plataformas y sus respectivas iniciales en el mismo orden
dataframes = [netflix_df, disney_df, hulu_df, amazon_df]
platforms = ['netflix', 'disney', 'hulu', 'amazon']
letters = ['n', 'd', 'h', 'a']
```

Transformacion sobre las fechas: De haber fechas, deberán tener el formato **`AAAA-mm-dd`**

```python


def date_format(dataframe: pd.DataFrame, date_column: str):
    """
    Cambia el formato de la fecha en una columna especificada de un DataFrame de "nombre del mes dd, yyyy" a "yyyy-mm-dd"
    """
    # Remover espacios en blanco en las fechas
    dataframe[date_column] = dataframe[date_column].str.strip()
    # Cambiar el formato de la fecha
    dataframe[date_column] = pd.to_datetime(dataframe[date_column], format='%B %d, %Y')
    dataframe[date_column] = dataframe[date_column].dt.strftime('%Y-%m-%d')


# Aplicar la función date_format a cada DataFrame en la lista
for df in dataframes:
    date_format(df, 'date_added')
```

Creacion de la columna ID con los criterios que se nos pidieron

Generar campo **`id`**: Cada id se compondrá de la primera letra del nombre de la plataforma,
seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = **`as123`**)

```python
def add_id_column(dataframe: pd.DataFrame, letter: str):
    """
    Agrega una nueva columna "id" al principio del contenido de la columna "show_id" en un DataFrame.
    """
    dataframe["id"] = letter+dataframe["show_id"]


# Aplicar la función add_id_column a cada DataFrame en la lista con la letra correspondiente
for i, df in enumerate(dataframes):
    add_id_column(df, letters[i])
```

Transformacion sobre los valores nulos de la columna "Rating"
Los valores nulos del campo rating deberán reemplazarse por el string “**`G`**” (corresponde al maturity rating: “general for all audiences”


```python
def replace_nan(dataframe: pd.DataFrame, value: str, column: str):
    """
    Reemplaza los valores nulos en una columna especificada del DataFrame con un valor especificado.
    """
    dataframe[column].fillna(value=value, inplace=True)
    


# Aplicar la función replace_nan a cada DataFrame en la lista
for df in dataframes:
    replace_nan(df, 'g','rating')

# Aplicar la funcion replace_nan a la columna duration de cada DataFrame en la lista para una funcion posterior (dado que daba error y esta fue mi forma de corregirlo)
    replace_nan(df, '0 min','duration')
```

Los campos de texto deberán estar en **minúsculas**, sin excepciones

```python
def lowercase(dataframe: pd.DataFrame):
    """
    Convierte todos los campos de texto en minúsculas en un DataFrame.
    """
    for col in dataframe.select_dtypes(include=['object']):
        dataframe[col] = dataframe[col].str.lower()

dataframes = [lowercase(df) for df in dataframes]
```

El campo ***duration*** debe convertirse en dos campos: **`duration_int`** y **`duration_type`**.
El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

```python
def split_duration(dataframe: pd.DataFrame):
    # Extraer el número y la unidad de medida de la columna 'duration'    
    dataframe[['duration_int', 'duration_type']] = dataframe['duration'].str.extract(r'(\d+)\s*([a-zA-Z]+)')
    # Convertir la columna 'duration_int' a tipo 'int'
    dataframe['duration_int'] = dataframe['duration_int'].astype(int)
    return dataframe

# Aplicar la función a cada DataFrame en la lista 'dataframes'
dataframes = [split_duration(df) for df in dataframes]
```

Reemplazamos la palabra "seasons" por la palabra "season" en la columna "duration_type" para que tenga consistencia con los otros valores.

```python
def replace_str(dataframe: pd.DataFrame, old_value: str, new_value: str, column: str):
    """
    Reemplaza los valores indicados en una columna especificada del DataFrame con un valor especificado.
    """
    dataframe[column] = dataframe[column].replace(old_value, new_value)
    


# Aplicar la función replace_str a cada DataFrame en la lista
for df in dataframes:
    replace_str(df, 'seasons','season','duration_type')
```

Creamos la columna "platform", para facilitarnos las busquedas que fueron solicitadas (sé que hay otras formas pero esta es la forma que se me ocurrió)

```python
for df, platform in zip(dataframes, platforms):
    df["platform"] = platform
```
